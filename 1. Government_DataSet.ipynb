```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline  
```

```python
file_path = "CRIME_REVIEW_FOR_THE_MONTH_OF_JANUARY_2024.csv"
df = pd.read_csv("CRIME_REVIEW_FOR_THE_MONTH_OF_JANUARY_2024.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
df.info()
```

```python
numerical_cols = [
    "During the current year upto the end of month under review",
    "During the corresponding month of previous year",
    "During the previous month",
    "During the current month"
]

# Grouping by major crime categories (if applicable)
if "Major Head" in df.columns:
    major_crime_counts = df.groupby("Major Head")[numerical_cols].sum()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df[numerical_cols[0]].dropna(), bins=20, kde=True)
plt.title(f"Distribution of {numerical_cols[0]}")
plt.show()
```

The graph shows that the majority of cases occurred in the early part of the year, with very few cases later on.

```python
plt.figure(figsize=(6, 4))
sns.histplot(df[numerical_cols[1]].dropna(), bins=20, kde=True)
plt.title(f"Distribution of {numerical_cols[1]}")
plt.show()
```

The graph shows that most cases occurred in the corresponding month of the previous year, with very few cases recorded afterward.

```python
plt.figure(figsize=(6, 4))
sns.histplot(df[numerical_cols[2]].dropna(), bins=20, kde=True)
plt.title(f"Distribution of {numerical_cols[2]}")
plt.show()
```

The graph indicates that the majority of cases were concentrated in the previous month, with a sharp decline in numbers thereafter.

```python
plt.figure(figsize=(6, 4))
sns.histplot(df[numerical_cols[3]].dropna(), bins=20, kde=True)
plt.title(f"Distribution of {numerical_cols[3]}")
plt.show()
```

The graph shows that most cases occurred during the current month, with a steep decline in frequency as the values increase.

```python
plt.figure(figsize=(10, 8))
top_categories = df.groupby("Major Head")[numerical_cols[0]].sum().nlargest(20).index
filtered_df = df[df["Major Head"].isin(top_categories)]

sns.stripplot(data=filtered_df, y="Major Head", x=numerical_cols[0], jitter=True, orient="h")
plt.title(f"Dot Plot of {numerical_cols[0]}")
plt.xlabel(numerical_cols[0])
plt.ylabel("Major Head")
plt.show()
```
